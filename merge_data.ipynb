```python
import pandas as pd
import numpy as np
```

```python
poke_df = pd.read_csv('pokemon.csv')
evo_df = pd.read_csv('evolutions.csv')
```

```python
merged = poke_df.merge(evo_df, on="name", how="left")
```

```python
merged = merged.rename(columns={"base-eperience": "base-experience"})
```

```python
merged = merged[["id", "name", "height", "weight", "base-experience", "evolves-from", "hp", "attack", "defense", "special-attack", "special-defense", "speed", "primary-type", "secondary-type"]]
```

```python
ids = merged["id"]
```

```python
regions = np.where(ids <= 151, "Kanto", 
        np.where((ids > 151) & (ids <= 251), "Johto",
                 np.where((ids > 251) & (ids <= 386), "Hoenn", 
                          np.where((ids > 386) & (ids <= 493), "Sinnoh",
                                   np.where((ids > 493) & (ids <= 649), "Unova",
                                            np.where((ids > 649) & (ids <= 721), "Kalos",
                                                     np.where((ids > 721) & (ids <= 809), "Alola",
                                                              np.where((ids > 809) & (ids <= 905), "Galar",
                                                                       np.where((ids > 905), "Paldea", ids)))))))))
generations = np.where(ids <= 151, 1, 
        np.where((ids > 151) & (ids <= 251), 2,
                 np.where((ids > 251) & (ids <= 386), 3, 
                          np.where((ids > 386) & (ids <= 493), 4,
                                   np.where((ids > 493) & (ids <= 649), 5,
                                            np.where((ids > 649) & (ids <= 721), 6,
                                                     np.where((ids > 721) & (ids <= 809), 7,
                                                              np.where((ids > 809) & (ids <= 905), 8,
                                                                       np.where((ids > 905), 9, ids)))))))))
```

```python
merged["Region"] = regions
merged["Generation"] = generations
```
